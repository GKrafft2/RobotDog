```python
import cv2
import matplotlib.pyplot as plt

# Load the MP4 video file
cap = cv2.VideoCapture('video/train/salut-droite-gui.mp4')
```

```python
# Calculate the total number of frames in the video
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
print(total_frames)
```

```python
# Calculate the frame interval for selecting 5 frames
frame_interval = int((total_frames - 1) / 4)

# Initialize the list of selected frames
frames = []

# Iterate over the frames in the video
for i in range(5):
    # Calculate the index of the next selected frame
    frame_index = i * frame_interval
    
    # Set the position of the video to the next selected frame
    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_index)
    
    # Read the next selected frame from the video
    ret, frame = cap.read()
    
    # Add the selected frame to the list of frames
    frames.append(frame)
    
    # Increment the position of the video by 10 frames
    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_index + 2)

# Display the selected frames
for i, frame in enumerate(frames):
    # Convert the color format of the frame from BGR to RGB
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    # Display the frame in the console
    plt.figure()
    plt.imshow(frame)
    plt.title('Frame ' + str(i+1))


# Wait for the user to close the windows
cv2.waitKey(0)

# Release the video capture object and close all windows
cap.release()
cv2.destroyAllWindows()
```
